### Template matching using correlation

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

image = mpimg.imread('Rain.bmp')
gray= rgb2gray(image)
Image = gray

image2 = mpimg.imread('A-white.bmp')
gray2 = rgb2gray(image2)
Image2 = gray2
plt.imshow(Image, cmap='gray')
```

```python
F_Image2 = np.fft.fft2(Image2)

F_Image = np.fft.fft2(Image)
F_I = np.conj(F_Image)
```

```python
F_IImage2 = F_Image2*F_I
F_IImage2I = abs(np.fft.fftshift(np.fft.fft2(F_IImage2)))
plt.imshow(F_IImage2I, cmap = 'gray')
```
