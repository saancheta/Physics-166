### Simulation of an imaging device

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

image = mpimg.imread('VIP.bmp')
gray= rgb2gray(image)
Image = gray
```

```python
x = np.linspace(-1,1,128)
y = x
X,Y = np.meshgrid(x,y)
r = np.sqrt(X**2 + Y**2)
Circle = (r<1)
plt.imshow(Circle, cmap = 'gray')
```

```python
F_Image = np.fft.fft2(Image)
F_Circle = np.fft.fftshift(Circle)

FRA = F_Circle*F_Image
IRA = abs(np.fft.fft2(FRA))

plt.imshow(IRA, cmap = 'gray')
```

```python
x = np.linspace(-1,1,128)
y = x
X,Y = np.meshgrid(x,y)
r = np.sqrt(X**2 + Y**2)
Circle = (r<0.1)
plt.imshow(Circle, cmap = 'gray')
```

```python
F_Image = np.fft.fft2(Image)
F_Circle = np.fft.fftshift(Circle)

FRA = F_Circle*F_Image
IRA = abs(np.fft.fft2(FRA))

plt.imshow(IRA, cmap = 'gray')
```

# It can be seen from these two results of "VIP" images that as the aperture of the lens is decreased, the quality of the image decreases.
