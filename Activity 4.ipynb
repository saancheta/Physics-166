### Edge detection using the convolution integral

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as mpimg
from scipy.signal import convolve2d
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

image = mpimg.imread('VIP.bmp')
gray= rgb2gray(image)
Image = gray
plt.imshow(Image, cmap='gray')
```

```python
hori = np.array([[-1, -1, -1],
              [2, 2, 2],
              [-1, -1, -1]])

vert = np.array([[-1, 2, -1],
              [-1, 2, -1],
              [-1, 2, -1]])

spot = np.array([[-1, -1, -1],
              [-1, 8, -1],
              [-1, -1, -1]])
```

```python
def conv2(x, y, mode = 'same'):
    return np.rot90(convolve2d(np.rot90(x, 2), np.rot90(y, 2), mode=mode), 2)
```

```python
F_Image = np.fft.fft2(Image)
F_IHori = abs(conv2(Image,hori))
F_IVert = abs(conv2(Image,vert))
F_ISpot = abs(conv2(Image,spot))
```

### Convolution of the VIP image with a 3x3 matrix pattern (horizontal)

```python
plt.imshow(F_IHori, cmap ='gray')
```

### Convolution of the VIP image with a 3x3 matrix pattern (vertical)

```python
plt.imshow(F_IVert, cmap ='gray')
```

### Convolution of the VIP image with a 3x3 matrix pattern (spot)

```python
plt.imshow(F_ISpot, cmap ='gray')
```
